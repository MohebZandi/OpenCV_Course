```python
# Creating a function to show and close window

def show_one_image(image_file):

    cv2.imshow('image', image_file)

    while True:
        k = cv2.waitKey(0) & 0xFF

        if k != 27:
            print('If you want to close, please press ESC')

        if k == 27:
            print('\nESC Pressed, closing window...')
            cv2.destroyAllWindows()
            break
```

### --------------------------------------------------------

### MouseEventTypes


#### Enumerator

EVENT_MOUSEMOVE 
Python: cv.EVENT_MOUSEMOVE
indicates that the mouse pointer has moved over the window.

EVENT_LBUTTONDOWN 
Python: **cv.EVENT_LBUTTONDOWN**
indicates that the left mouse button is pressed.

EVENT_RBUTTONDOWN 
Python: cv.EVENT_RBUTTONDOWN
indicates that the right mouse button is pressed.

EVENT_MBUTTONDOWN 
Python: cv.EVENT_MBUTTONDOWN
indicates that the middle mouse button is pressed.

EVENT_LBUTTONUP 
Python: **cv.EVENT_LBUTTONUP**
indicates that left mouse button is released.

EVENT_RBUTTONUP 
Python: cv.EVENT_RBUTTONUP
indicates that right mouse button is released.

EVENT_MBUTTONUP 
Python: cv.EVENT_MBUTTONUP
indicates that middle mouse button is released.

EVENT_LBUTTONDBLCLK 
Python: **cv.EVENT_LBUTTONDBLCLK**
indicates that left mouse button is double clicked.

EVENT_RBUTTONDBLCLK 
Python: cv.EVENT_RBUTTONDBLCLK
indicates that right mouse button is double clicked.

EVENT_MBUTTONDBLCLK 
Python: cv.EVENT_MBUTTONDBLCLK
indicates that middle mouse button is double clicked.

EVENT_MOUSEWHEEL 
Python: cv.EVENT_MOUSEWHEEL
positive and negative values mean forward and backward scrolling, respectively.

EVENT_MOUSEHWHEEL 
Python: cv.EVENT_MOUSEHWHEEL
positive and negative values mean right and left scrolling, respectively.

### --------------------------------------------------------------

### MouseEventFlags

#### Enumerator

EVENT_FLAG_LBUTTON 
Python: cv.EVENT_FLAG_LBUTTON

## ترسیم دایره روی تصویر خالی

```python
# Create a black image
img = np.zeros((512,512,3), np.uint8)
```

```python
show_one_image(img)
```

```python
# Create a function based on a CV2 Event (Left button click)
def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN: 
        cv2.circle(img,(x,y),20,(0,255,0),-1)

img = np.zeros((512,512,3), np.uint8)

# This names the window so we can reference it to mouse event
cv2.namedWindow(winname = 'my_drawing')

# Connects the mouse button to our callback function
cv2.setMouseCallback('my_drawing',draw_circle)

while True: 
    # Shows the image window
    cv2.imshow('my_drawing',img)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```

```python
# Create a function based on a CV2 Event (Left button click) and CLEAR image by prsseing 'c' key

def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN: 
        cv2.circle(img,(x,y),20,(0,255,0),-1)

img = np.zeros((512,512,3), np.uint8)

# This names the window so we can reference it to mouse event
cv2.namedWindow(winname = 'my_drawing')

# Connects the mouse button to our callback function
cv2.setMouseCallback('my_drawing',draw_circle)

while True: 
    # Shows the image window
    cv2.imshow('my_drawing',img)
   
    # Clear the image from circles
    if cv2.waitKey(10) & 0xFF == ord('c'):
        img = np.zeros((512,512,3), np.uint8) 

    elif cv2.waitKey(1) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```

### Drawing a rectangle on Image

```python

# Defining variables to use in program
drawing = False
point1 = ()
point2 = ()

def Mouse_drawing(event, x, y, flags, params):
    global point1, point2, drawing
    
    if event == cv2.EVENT_LBUTTONDOWN:
        if drawing == False:
            drawing = True
            point1 = (x, y)
        else:
            drawing = False
            
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            point2 = (x, y)
            
        
# Reading the Image
# img = cv2.imread('../images/cat.jpg')
img = np.zeros((512,512,3), np.uint8)
# Define Window name
cv2.namedWindow('Image')

# Connect mouse function to callback
cv2.setMouseCallback('Image', Mouse_drawing)

# Continious showing the image
while True:
    cv2.imshow('Image', img)
    
    if point1 and point2:
#         img = cv2.imread('../images/cat.jpg')
        img = np.zeros((512,512,3), np.uint8)
        cv2.rectangle(img, point1, point2, (0, 255, 0), 3)
           
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break

cv2.destroyAllWindows()
```

### Drawing a rectangle on Video

```python
# Drawing on a Video


# Defining variables to use in program
drawing = False
point1 = ()
point2 = ()

def get_points_mouse(event, x, y, flags, params):
    global point1, point2, drawing
    
    if event == cv2.EVENT_LBUTTONDOWN:
        if drawing is False:
            drawing = True
            point1 = (x, y)
        else:
            drawing = False
            
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            point2 = (x, y)
            
        
# Reading the Image
video_file = '../images/Sponge_Bob.mp4'
vid = cv2.VideoCapture(video_file)

# Define Window name
cv2.namedWindow('Image', cv2.WINDOW_NORMAL)

# Connect mouse function to callback
cv2.setMouseCallback('Image', get_points_mouse)

# Continious showing the image
while True:
    _, frame = vid.read()
    
    if point1 and point2:
        cv2.rectangle(frame, point1, point2, (0, 0, 255), 3)
    
    cv2.imshow('Image', frame)
    k = cv2.waitKey(20) & 0xFF
    if k == 27:
        break

cv2.destroyAllWindows()
```

## Drawing a circle on Video

### calculate radius from two points (x1,y1) and (x2,y2) ---> distance of points

<img src='../images/distance.png' width=400>

```python
# Drawing on a Video


# Defining variables to use in program
drawing = False
point1 = ()
point2 = ()

def get_points_mouse(event, x, y, flags, params):
    global point1, point2, drawing
    
    if event == cv2.EVENT_LBUTTONDOWN:
        if drawing is False:
            drawing = True
            point1 = (x, y)
        else:
            drawing = False
            
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            point2 = (x, y)
            
        
# Reading the Image
video_file = '../images/Car_Drift.mp4'
vid = cv2.VideoCapture(video_file)

# Define Window name
cv2.namedWindow('Image')

# Connect mouse function to callback
cv2.setMouseCallback('Image', get_points_mouse)

# Continious showing the image
while True:
    ret, frame = vid.read()
    if ret:
        if point1 and point2:
            # calculate radius from two points (x1,y1) and (x2,y2)
            r = int(np.sqrt(np.square(point1[0] - point2[0]) + np.square(point1[1] - point2[1])))
            cv2.circle(frame, point1, r, (255, 0, 255), 3)

        cv2.imshow('Image', frame)
        k = cv2.waitKey(20) & 0xFF
        if k == 27:
            break
    else:
        break

cv2.destroyAllWindows()
```
