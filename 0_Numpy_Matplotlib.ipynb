#### Showing images

```python
image = np.random.randint(0, 255, dtype= 'uint8')
image
```

```python
image = np.zeros((100,100), dtype='uint8')
image
```

```python
plt.imshow(image, cmap= 'gray')
```

```python
for i in range(100):
    for j in range(100):
        if i % 2 == 0:
            image[i,j] = np.random.randint(0, 255, dtype= 'uint8')
image
```

```python
plt.imshow(image)
```

```python
fig = plt.figure(figsize= (8,12))
plt.imshow(image, cmap='gray')
```

```python
plt.imshow(image[:30,25:35], cmap='gray')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mat = np.arange(0,10)
mat
```

```python
plt.plot(mat)
```

```python
mat2 = mat.copy()
mat2 = sorted(mat2, reverse=True)
mat2
```

```python
plt.plot(mat)
plt.plot(mat2)
plt.show()
```

```python
plt.plot(mat)
plt.plot(mat2)
plt.legend(['mat', 'mat2'])
plt.show()
```

####  Creating Subplot

```python
plt.subplot(1,2,1)
plt.plot(mat)

plt.subplot(1,2,2)
plt.plot(mat2, color='red')
plt.show()
```

```python
plt.subplot(211)
plt.plot(mat)

plt.subplot(212)
plt.plot(mat2, color='red')
plt.show()
```

```python
# Creating subplots by plt.subplots()

# TAKE CARE OF S AT THE END OF SUBPLOTS

fig, ax = plt.subplots(1,2)     
ax[0].plot(mat)
ax[1].plot(mat2, color= 'r')
```

```python
# Creating subplots by plt.subplots()

fig, ax = plt.subplots(1,2)     # TAKE CARE OF S AT THE END OF SUBPLOTS
ax[0].plot(mat)
ax[1].plot(mat2, color= 'r')

# Add a title for total charts
fig.suptitle('My Two Charts Tilte')

# Add a title for any of the charts
ax[0].set_title('My Charts')

# Turn Axis Off
ax[0].axis('off')
```

```python
# Adding label to axes

fig, ax = plt.subplots(1,2)     
ax[0].plot(mat)
ax[1].plot(mat2, color= 'r')

# Adding label to axes
ax[1].set_xlabel('time (s)')
ax[1].set_ylabel('Count (Number)', color = 'green' )
```

```python
# More columns and rows of subplots

fig, ax = plt.subplots(2,3)     
ax[0,1].plot(mat)
ax[1,0].plot(mat2, color= 'r')
```

```python
fig, ax = plt.subplots(1,2)     # , sharey=True) 
ax[0].plot(mat[:3])
ax[1].plot(mat2, color= 'r')
```
