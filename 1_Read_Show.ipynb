#### اگر در برنامه بالا کلیدی غیر از کلید مشخص شده فشرده شود برنامه هنگ می کند

#### برای برطرف کردن این مشکل از یک حلقه بینهایت استفاده می کنیم که اگر کلید دیگری فشرده شد برنامه هنگ نکند و با فشرده شدن کلید مناسب پنجره بسته شود

```python
cv2.imshow('my window', img)

while True:
    k = cv2.waitKey(0)
    
    if k != 27:
        print(k)
        print('If you want to close, please press ESC')
    
    if k == 27:
        print('ESC Pressed, closing window...')
        print(k)
        cv2.destroyAllWindows()
        break
```

```python
filename_video = '../images/slow_traffic_small.mp4'
vid = cv2.VideoCapture(filename_video)

while True:
    ret, frame = vid.read()
    print (ret)
    if ret:
        cv2.imshow('Video', frame)

        k = cv2.waitKey(30) & 0xFF   # Make a delayto show the video and get the ESC key
        if k == 27 :    
            break
    else:
        break

vid.release()
cv2.destroyAllWindows()
```

**0xFF** is a hexadecimal constant which is 11111111 in binary. 

By using bitwise AND ( & ) with this constant, it leaves only the last 8 bits of the original (in this case, whatever cv2. waitKey(0) is).

### Reading from Camera

```python
vid = cv2.VideoCapture(0)

while True:
    ret, frame = vid.read()

    if ret:
        cv2.imshow('Video', frame)

        k = cv2.waitKey(25) & 0xFF
        if k == 27:
            break
    else:
        break

vid.release()
cv2.destroyAllWindows()
```

```python
# Make the video from camera grayscale

vid = cv2.VideoCapture(0)

while True:
    ret, frame = vid.read()

    if ret:
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow('Video', frame)

        k = cv2.waitKey(25) & 0xFF
        if k == 27:
            break
    else:
        break

vid.release()
cv2.destroyAllWindows()
```

#### یک فایل تصویر را بخوانید 

#### ابعاد و تعداد کانال فایل خوانده شده را بیابید

#### فایل خوانده شده را نمایش دهید بطوریکه با فشردن کلید مشخصی مثلا اسکیپ بسته شود

#### فایل تصویر را خوانده و تصویر رنگی و خاکستری را نمایش و با فشردن کلید کیو تصویر اصلی و با فشردن کلید اسکیپ همه تصاویر بسته شوند

```python
import cv2
print(' Press "q" to close gray image, Press ESC to close all windows')
filename0 = 'squirrel_cls.jpg'
filename1 = 'fruits.jpg'

img = cv2.imread(image_path + filename0)
img_gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)

cv2.namedWindow('My Original image',cv2.WINDOW_AUTOSIZE)
cv2.namedWindow('My gray image',cv2.WINDOW_AUTOSIZE)

cv2.imshow('My Original image', img)
cv2.imshow('My gray image', img_gray)

while True:
    k = cv2.waitKey(0)
    
    if k != ord('q') or k != 27:
        print(' Press "q" to close gray image, Press ESC to close all windows')
        
    if k == ord('q'): 
        cv2.destroyWindow('My gray image')
        print('"q" pressed, Gray image closed')
        print('If you want to close, please press ESC')
    
    if k == 27:
        print('\nESC Pressed, closing window...')
        cv2.destroyAllWindows()
        break
```
